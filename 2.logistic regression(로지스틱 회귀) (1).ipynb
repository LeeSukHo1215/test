# 2.logistic regression(로지스틱 회귀)

##정의
*  독립 변수의 선형 결합을 이용하여 사건의 발생 가능성을 예측하는데 사용되는 통계 기법 

## 설명

예를 들어 
* 고양이 사진인지 아닌지 구분하고 싶은 사진이 주어졌을 때 y의 예측값을 출력하는 알고리즘을 원합니다
* 더 자세히는 y의 예측값은 입력 feature x가 주어졌을 때 y가 1일 확률을 뜻합니다

* 다른 말로 x가 지난 시간에 본 사진이라면
y의 예측값은 이 사진이 고양이 사진일 확률을 알려줍니다


파라미터 w와 b는 각각 차원 상의 벡터와 실수입니다


### 틀린 예시
y의 예측값을
w의 전치 × x + b로 놓는다면
$$\hat{y} = w^{T}x+ b$$
y의 예측값은 y가 1일 확률이기 때문에
항상 0과 1사이어야하는데
w의 전치 × x + b는 1보다 훨씬 크거나
음수일 수도 있어서
0과 1 사이어야 하는 확률의 값으로 성립하지 않는다

#### 시그모이드 함수의 공식
1) z가 실수일 때 z의 시그모이드는
$$\sigma (z) = \frac{1}{1+e^{-z}}$$   

2) z 값이 매우 크다면
$$ \sigma (z) \approx \frac{1}{1+0} \approx 1 $$

3) z 값이 매우 작다면 
$$ \sigma (z) \approx \frac{1}{1+\infty} \approx 0 $$
